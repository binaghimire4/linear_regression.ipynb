```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
df = pd.read_csv(r"C:\data for university\data_science_salaries.csv")
print(df.info())
print(df.describe())
```

```python
#Features and Target
features = ['yearsofexperience', 'basesalary', 'bonus', 'stockgrantvalue']

X = df [features]
y = df ['totalyearlycompensation']
```

```python
#fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
#Prediction
y_pred = model.predict(X)
```

```python
#Matrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print results
print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

```python
MSE is the avearage of the squares of the error difference between the prediction and actual value. 
The lower the MSE, the better the model is at predicting the target variable.

R2 is 76.8% of the variance in the target variable can be explained by the features. The closer R2 to 1, the better the model explain the data.
```
